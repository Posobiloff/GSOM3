**1. A list of tickets, where number of flights>3**

```python
%%sql $CONNECT_DATA
SELECT ticket_no 
FROM Ticket_flights 
GROUP BY ticket_no
HAVING COUNT(flight_id) > 3 LIMIT 1000
```

**2. Output a total number of tickets which have exactly 5 flights. Just one number, nothing else.**

```python
%%sql $CONNECT_DATA
SELECT COUNT(t.ticket_no) as number_of_tickets_with_5_flights
FROM (SELECT ticket_no
      FROM Ticket_flights
      GROUP BY ticket_no
      HAVING COUNT(flight_id) = 5) as t
```

**4. the Department_name which has the longest-working employee in the company, together with the First_name + Last_name of this employee**

```python
%%sql postgresql:///jovyan
SELECT First_name, Last_name, Department_name 
from Employees e
JOIN Departments d ON
e.DEPARTMENT_ID = d.DEPARTMENT_ID
WHERE HIRE_DATE = (SELECT MIN(HIRE_DATE) 
from Employees e)
```
