```python
#Solution with subqueries
```

```python
%%sql postgresql:///jovyan
SELECT department_name, CONCAT(first_name,' ',last_name) as Manager_name, Number_of_employees
FROM Employees e
JOIN (SELECT Department_name, COUNT(employee_id) as Number_of_employees, d.manager_id
     FROM Departments d
      JOIN Employees USING (department_id) 
      GROUP BY Department_name, d.manager_id
     ORDER BY Department_name) as k
ON e.employee_id = k.manager_id
```

```python
#Solution without subqueries
```

```python
%%sql postgresql:///jovyan
SELECT department_name, CONCAT(e2.first_name,' ',e2.last_name) as Manager_name, COUNT(e1.employee_id) as Number_of_employees
FROM Departments d
JOIN Employees e1 ON e1.department_id = d.department_id
JOIN Employees e2 ON e2.employee_id = d.manager_id
GROUP BY d.department_name, e2.first_name, e2.last_name
ORDER BY d.department_name
```
